```python
def split_point(point):
    if point:
        try:
            point_split = point[1:-1].split(",")
            X = float(point_split[0])
            Y = float(point_split[1])
            return (X, Y)
        except:
            return (0.0,0.0)
    else:
        return (0.0,0.0)

def get_X(point):
    return split_point(point['Point'])[1]

def get_Y(point):
    return split_point(point['Point'])[0]

cs['X'] = cs.apply(get_X, axis=1)
cs['Y'] = cs.apply(get_Y, axis=1)
```

```python
cs.head()
```

```python
# 1 ft corresponds to 0.000002746181435 degree shift in lat/long
# 1 block is roughly 500 ft in SF
# 1 block is roughly 0.001373090717 degree shift in both lat/long
import math, datetime
block_norm = 0.001373090717*2
time_block = 600 # 10 mins

def normalize_lat(lat):
    lat_blocks = float(lat-(-123))/block_norm
    return math.floor(lat_blocks)

def normalize_long(longt):
    longt_blocks = float(longt-37)/block_norm
    return math.floor(longt_blocks)

def add_lat_long_blocks(row):
    return str(int(normalize_lat(float(row['X']))))+','+str(int(normalize_long(float(row['Y']))))

def normalize_datetime(row):
    try:
        datetime_row = row['DateTime']
        a = (datetime_row - datetime.datetime(1970,1,1)).total_seconds()
        five_blocks = math.floor(float(a)/float(time_block))
        return five_blocks
    except:
        return 0.0

df['time_vicinity'] = df.apply(normalize_datetime, axis=1)
cs['time_vicinity'] = cs.apply(normalize_datetime, axis=1)
df['block_vicinity'] = df.apply(add_lat_long_blocks, axis=1)
cs['block_vicinity'] = cs.apply(add_lat_long_blocks, axis=1)
```

```python
def add_latlng_datetime(row):
    return str(row['time_vicinity'])+' '+str(row['block_vicinity'])

df['latlng_datetime'] = df.apply(add_latlng_datetime, axis=1)
cs['latlng_datetime'] = cs.apply(add_latlng_datetime, axis=1)
```

1. Use address - tokenizing
2. Use lat and long by setting up a perimeter (to test with multiple parameters) - seems most promising
3. Use date and time with a window period
4. Use combination of date + time and lat + long with a larger window period for each one
5. Use combination of just time with lat + long with larger window periods for lat/long but smaller ones for time

With fuzzy matches, we're likely to have multiple matches. There are 2 things to consider

1. Should we reduce our 'fuzziness' so that it reduces to maximum number of single matches?
2. Is there value in having multiple matches?
3. More matches have a higher predictive power
